```python
def plot_setup(plot_type):
	# bigger figure for presentations:
	if plot_type == 'presentation':
		fig = plt.figure(figsize = (25,18))
		# fontsize = 'big'
		fs = 45
		ts = 20
		wt = 2.5
	elif plot_type == 'paper':
		# medium size for paper, with big labels:
		fig = plt.figure(figsize = (12,9))	
		# fontsize = 'medium'	
		fs = 40
		ts = 10
		wt = 1.5
	elif plot_type == 'notebook':
		fig = plt.figure(figsize = (11,7))	
		# fontsize = 'normal'			
		fs = 25
		ts = 10
		wt = 1.5		

	plt.rc('font',**{'family':'serif','serif':['Helvetica'], 'weight' : 'normal', 'size' : fs})	

	plt.rc('text', usetex=True)
	plt.rc('axes', linewidth=2)
		
	#if fontsize == 'big':
	plt.rc('xtick', labelsize=fs-5)
	plt.rc('xtick.major', size=ts, width=wt)
	plt.rc('xtick.minor', size=7, width=1.0)

	plt.rc('ytick', labelsize=fs-5)
	plt.rc('ytick.major', size=ts, width=wt)
	plt.rc('ytick.minor', size=7, width=1.0)
		
	return
```

```python
# setup

#log-spaced mass in units of M_sun/h
log_mass_min = 9.
log_mass_max = 16.
nmass = 5
mass = np.logspace(log_mass_min, log_mass_max, nmass)

# z vector
zmin = 0.0
zmax = 1.0
nz = 3
z = np.linspace(zmin, zmax, nz)

# log-spaced k vector
kmin = 0.1
kmax = 1000.
nk = 1000
k = np.logspace(np.log10(kmin), np.log10(kmax), nk)
```

```python
import hankel

def uell_nfw_test(nfw_profile, k, z, r_s, rvir, c, mass, ell):
    msize = mass.size #c.shape[1]
    mnfw = mass_nfw(r_s, c)
    print "c.size, k.size, z.size, msize"
    print c.size, k.size, z.size, msize
    uk_l = np.zeros((z.size, msize, k.size))
    
    h_highmass = hankel.hankel.HankelTransform(ell+0.5,1000,0.00005) #perfect for im=3,4
    h_lowmass = hankel.hankel.HankelTransform(ell+0.5,10000,0.00000005) #ok for im=1
    
    #h = hankel.hankel.HankelTransform(ell+0.5,900,0.00005) #perfect for im=3,4
    #h = hankel.hankel.HankelTransform(ell+0.5,10000,0.00000005) #ok for im=1
    #h = hankel.hankel.HankelTransform(ell+0.5,100000,0.000000005) #perfect for im=3,4
    for jz in range(z.size):
        for im in range(msize):
            for kk in range(k.size):
                nfw_f = lambda x: nfw_profile(x/k[kk], r_s[jz,im])*(x**2.)*np.sqrt(np.pi/(2.*x))*vector_step_function(x/k[kk], rvir[im])
                if mass[im]<=1e12:
                    uk_l[jz, im, kk] = h_lowmass.integrate(nfw_f)[0]/(k[kk]**3.*mnfw[jz,im])
                else:
                    uk_l[jz, im, kk] = h_highmass.integrate(nfw_f)[0]/(k[kk]**3.*mnfw[jz,im])
    return uk_l


#------------------------------ uell -----------------------------------#

# create a 4dim array containing u_ell as a function of l,z,m and k
# uell[l,z,m,k]
def IA_uell_hankel_test(k, c, z, r_s, rvir, mass, ell_max):
    msize = rvir.size #c.shape[1]
    uell = np.zeros((ell_max+1, z.size, msize, k.size)) 
    for i in range(0,ell_max/2):
            uell[2*i] = uell_nfw_test(nfw_profile, k, z, r_s, rvir, c, mass, 2*i)
            #uell[2*i+1] = 0
    return uell
```

```python
# execute

#---- Comology ----#
# Load cosmological parameters (here SB10)
this_cosmo = FlatLambdaCDM(H0=71.9, Om0=0.3)
rho_crit0 = this_cosmo.critical_density0.to(u.M_sun*u.Mpc**(-3.))/(this_cosmo.h**2.)
omega_m = this_cosmo.Om0

#concentration -> to be replaced by a module
c = np.empty([nz,nmass])
mass_term = 10.14/((mass/(2.e+12))**0.081)
for jz in range (0,nz):
    for im in range (0,nmass):
        #c[jz,im]= 10.14*mass_term[im]/((1.+z[jz])**1.01) #Duffy
        c[jz,im]= mass_term[im]/((1.+z[jz])**1.01) #Cooray&Sheth 2002 (used in SB10)


# compute the virial radius and the scale radius associated with a halo of mass M
rho_m = minimal_cosmo(omega_m)
rho_halo = 200.*rho_m
rvir = radvir(mass, rho_halo)
r_s = np.empty([nz, nmass])
for jz in range(0,nz):
    r_s[jz] = rvir/c[jz]

print("c.shape = ", c.shape)
print("rvir.shape = ", rvir.shape)
print("r_s.shape = ", r_s.shape) 

ell_max = 8

# uell(l,z,m,k)
uell = IA_uell_hankel_test(k, c, z, r_s, rvir, mass, ell_max)

print (uell.shape)
```

```python
# compare with the analytical one:
from scipy.special import sici


#Navarro-Frenk-White density profile
def norm_fourier(x, c):
    # Note: x = k*r_scale where r_scale= r200/c
    si, ci = sici((1+c)*x)
    si2, ci2 = sici(x)
    sinterm = np.sin(x)*( si-si2 ) -np.sin(c*x)/((1+c)*x)
    costerm = np.cos(x)*( ci-ci2 )
    f_c = np.log(1+c)-c/(1+c)
    u_fourier = ( sinterm + costerm )/f_c
    return u_fourier

uk_anyc = np.empty([nz, nmass, len(k)])
for jz in range(0,nz):
    for im in range(0,nmass):
        #for i in range(0,nk):
        uk_anyc[jz,im] = norm_fourier(r_s[jz,im]*k, c[jz,im])
```

```python
#------------------------------ w(k|m) ---------------------------------#

#integral of the angular part in eq B8 (SB10) using the Legendre polynomials
#assuming theta_e=theta, phi_e=phi (perfect radial alignment) and theta_k = pi/2

from numpy import pi

def pl():
    p = np.empty(7)
    p[0] = p[1] = p[3] = p[5] = 0
    p[2] = (9./32)*pi**2.
    p[4] = (-10./512.)*pi**2.
    p[6] = (19530./524288.)*pi**2.
    return p

def w_km(lmax, uell):
    p = pl()
    w_l = 0
    for ll in range (0,lmax/2+1):
        l = 2*ll
        w_l += (2*l+1.)*((1j)**l)*uell[l]*p[l]
        print (2*l+1.)*((1j)**l)
    wk = np.absolute(w_l)
    return wk 
```

```python
import numpy as np
import hankel
import math

import time
from itertools import count, izip
from numpy import absolute, array, arange, cos, exp, linspace, log10, \
                  logspace, max as npmax, median, newaxis, ones, outer,\
                  pi, sin,  sum as npsum, zeros
from scipy.integrate import quad, simps, trapz
from scipy.interpolate import interp1d
from scipy.misc import factorial
from scipy.special import legendre, sici, binom
import math

import matplotlib.pyplot as plt
%matplotlib inline
```

```python



NewtonGravConst = 4.2994E-9
rho_crit0= (3.*10**4)/(8.*math.pi*NewtonGravConst) 		#critical density at redshift 0


#-----------------------------------------------------------------------#
#								u_ell									#
#-----------------------------------------------------------------------#

#Since we are interested on the normalised nfw profile only, 
#here all the constant have been suppressed (rho_s = rho_mean*delta_char)
'''
def nfw_profile(r, r_s):
	#rho_nfw(c,r) -> it is a two dim array, in c and r
	rho_nfw = np.zeros((r_s.size,r.size))
	for j in range(r_s.size):
		x = r/r_s[j]
		f_x = x*(1.+x)**2
		rho_nfw[j] = 1./f_x

	return rho_nfw
'''	
def nfw_profile_v1(r, r_s, i,j):
	x = r/r_s[i,j]
	f_x = x*(1.+x)**2
	rho_nfw = 1./f_x
	return rho_nfw
	
def nfw_profile(r, rs):
	x = r/rs
	f_x = x*(1.+x)**2
	rho_nfw = 1./f_x
	return rho_nfw
	
def step_function(x, treshold):
	if x.size >1:
		y = np.zeros(x.size)
		for i in range(x.size):
			if x[i]<treshold:
				y[i] = 1.
	else:
		if x<treshold:
			y = 1.
		else:
			y = 0
	return y

def scalar_step_function(x, threshold):
	print 'x = ', x
	print 'threshold = ', threshold
	if x<threshold:
		y = 1.
	else:
		y = 0
	return y
	
def vector_step_function(x, threshold):
	#print 'x.shape = ', x.shape
	mask_x = x<threshold
	y = np.zeros(x.shape)
	y[mask_x] = 1.
	return y	


#Navarro-Frenk-White density profile
def norm_fourier(x, c):
	# Note: x = k*r_scale where r_scale= r200/c
	si, ci = sici((1+c)*x)
	si2, ci2 = sici(x)
	sinterm = np.sin(x)*( si-si2 ) -np.sin(c*x)/((1+c)*x)
	costerm = np.cos(x)*( ci-ci2 )
	f_c = np.log(1+c)-c/(1+c)
	u_fourier = ( sinterm + costerm )/f_c
	return u_fourier


def minimal_cosmo(Omega_m):
	rho_m = rho_crit0*Omega_m 						#comoving average matter density of the Universe	
	return rho_m
	
def radvir(m, rho_halo):
	radvir_constants = 3./(4.*math.pi*rho_halo) 	
	r_vir = (m*radvir_constants)**(1./3.) #Mpc/h	
	return r_vir


#ht = hankel.SphericalHankelTransform(nu= 0,		# The order of the bessel function
#                     N = 1000,          			# Number of steps in the integration
#                     h = 0.03)						# Proxy for "size" of steps in integration


def mass_nfw(r_s, c):
	mnfw = (r_s**3.)*(np.log(1.+c)-c/(1.+c))
	print 'mnfw.shape = ', mnfw.shape
	return mnfw


from scipy.interpolate import interp1d

def uell_nfw(nfw_profile, k, z, r_s, rvir, c, mass, ell):
    msize = mass.size #c.shape[1]
    mnfw = mass_nfw(r_s, c)
    print "c.size, k.size, z.size, msize"
    print c.size, k.size, z.size, msize
    uk_l = np.zeros((z.size, msize, k.size))
    
    # see the jupyter notebook wkm_module_tests for convergence and ringing tests
    h_highmass = hankel.hankel.HankelTransform(ell+0.5,1000,0.00005) 
    h_lowmass = hankel.hankel.HankelTransform(ell+0.5,10000,0.00000005) 

    for jz in range(z.size):
        for im in range(msize):
            for kk in range(k.size):
                nfw_f = lambda x: nfw_profile(x/k[kk], r_s[jz,im])*(x**2.)*np.sqrt(np.pi/(2.*x))*vector_step_function(x/k[kk], rvir[im])
                if mass[im]<=1e12:
                    uk_l[jz, im, kk] = h_lowmass.integrate(nfw_f)[0]/(k[kk]**3.*mnfw[jz,im])
                else:
                    uk_l[jz, im, kk] = h_highmass.integrate(nfw_f)[0]/(k[kk]**3.*mnfw[jz,im])
    return uk_l


#------------------------------ uell -----------------------------------#

# create a 4dim array containing u_ell as a function of l,z,m and k
# uell[l,z,m,k]
def IA_uell_hankel(k, c, z, r_s, rvir, mass, ell_max):
    msize = rvir.size #c.shape[1]
    uell = np.zeros((ell_max+1, z.size, msize, k.size)) 
    #for i in range(0,ell_max/2+1):
    #    uell[2*i] = uell_nfw(nfw_profile, k, z, r_s, rvir, c, mass, 2*i)
    #    #uell[2*i+1] = 0
    for i in range(0,ell_max+1):
        uell[i] = uell_nfw(nfw_profile, k, z, r_s, rvir, c, mass, i)
        #uell[2*i+1] = 0
    return uell
```

```python
def wkm_radial_part(uell, ell_max):
    nl = np.size(uell, axis=0)
    nz = np.size(uell, axis=1)
    nm = np.size(uell, axis=2)
    nk = np.size(uell, axis=3)
    print 'nl, nz, nm, nk = ', nl, nz, nm, nk
    sum_ell = np.zeros([nz,nm,nk])
    for jz in range(nz):
        for im in range(nm):
            for ell in range(0,ell_max+1):
                sum_ell[jz,im] = sum_ell[jz,im] + (1j)**(ell) * (2.*ell + 1.) * uell[ell,jz,im]
    return sum_ell
```

```python
def f2_analytic(theta_k, phi_k):
    return 9./32. * np.pi**2. * np.sin(theta_k)**2. * ( np.cos(2*phi_k) + 1j * np.sin(2.*phi_k) ) 
```

```python
from wkm_angular_part import g
def f4_analytic(theta_k, phi_k):
    ei2phik = np.cos(2.*phi_k) + 1j*np.sin(2.*phi_k)
    term1 = 175./64. * np.pi * np.sin(theta_k)**4. * g(4) * ei2phik
    term4 = 105./32. * np.pi * np.sin(theta_k)**2. * np.cos(theta_k)**2. * g(2) * ei2phik
    #term1 = 175./128. * np.pi * np.sin(theta_k)**4. * g(4) * ei2phik
    #term3 = 70./256. * np.pi * np.sin(theta_k)**2. * np.cos(theta_k)**2. * g(2) * ei2phik
    #term4 = 35./64. * np.pi * np.sin(theta_k)**2. * np.cos(theta_k)**2. * g(2) * ei2phik
    term7 = -45./32. * np.pi * np.sin(theta_k)**2. * g(2) * ei2phik
    #return term1 + term3 + term4 + term7
    return term1/2. + term4 + term7
```

```python
def fell_analytic(theta_k, phi_k, ell_max):
    ei2phik = np.cos(2.*phi_k) + 1j*np.sin(2.*phi_k)
    sink = np.sin(theta_k)
    cosk = np.cos(theta_k)
    pi = np.pi
    
    f_ell=0
    
    x2 = 3./8. * pi * sink**2. * g(2) * ei2phik
    x4 = ( 5./16. * pi * sink**4. * g(4) + 3./8. * pi * sink**2. * cosk**2. * g(2) ) * ei2phik
    
    if ell_max == 2:
        f_ell = 3./2. * x2
    if ell_max == 4:
        f_ell = 35./8. * x4 - 30./8. * x2
    return f_ell
```

```python
def another_fell(theta_k, phi_k, l):
	phase = np.cos(2.*phi_k) + 1j*np.sin(2.*phi_k)
	sum1 = 0.
	for m in range(0,l+1):
		#print 'm = ', m
		#print 'newton_bin(l,m)*newton_bin(l+m-1,l) = ', newton_bin(l,m)*newton_bin(l+m-1,l)
		sum2=0.
		#print 'my coefficients = ', newton_bin(l,m)*newton_bin(l+m-1,l)
		for j in range(0,m+1):
			sum2 += binom(m,j) * g(m-j) * np.sin(theta_k)**(m-j) * np.cos(theta_k)**j * I_x(m-j+1,j)
			#print 'j = ', j
			#print 'sum2 = ', sum2
		sum1 += binom(l,m)*binom(0.5*(l+m-1.),l)*sum2		
	return 2.**l * sum1*phase 
```

```python
def wkm_test(uell, theta_k, phi_k, ell_max):
    nl = np.size(uell, axis=0)
    nz = np.size(uell, axis=1)
    nm = np.size(uell, axis=2)
    nk = np.size(uell, axis=3)
    #print 'nl, nz, nm, nk = ', nl, nz, nm, nk
    sum_ell = np.zeros([nz,nm,nk])
    for jz in range(nz):
        for im in range(nm):
            for ell in range(0,ell_max+1):
                radial = (1j)**(ell) * (2.*ell + 1.) * uell[ell,jz,im]
                angular = fell_analytic(theta_k, phi_k, ell)
                #print ell, angular
                sum_ell[jz,im] = sum_ell[jz,im] + radial*angular
                #print sum_ell
    return sum_ell
```

```python
def wkm_my_fell(uell, theta_k, phi_k, ell_max):
    nl = np.size(uell, axis=0)
    nz = np.size(uell, axis=1)
    nm = np.size(uell, axis=2)
    nk = np.size(uell, axis=3)
    #print 'nl, nz, nm, nk = ', nl, nz, nm, nk
    sum_ell = np.zeros([nz,nm,nk])
    for jz in range(nz):
        for im in range(nm):
            for ell in range(0,ell_max+1):
                radial = (1j)**(ell) * (2.*ell + 1.) * uell[ell,jz,im]
                angular = another_fell_phik(theta_k, phi_k, ell)
                #print ell, angular
                sum_ell[jz,im] = sum_ell[jz,im] + radial*angular
                #print sum_ell
    return np.sqrt(np.real(sum_ell)**2.+np.imag(sum_ell)**2.)
```
