We adopt the following convention to define (ra,dec) coordinates from cartesian (x,y,z)
ra = 180/pi*atan(x,y) and dec= 90-180/pi*atan(sqrt(x*x+y*y),z) where (ra,dec) are given in
degrees, and atan(x,y) = atan(x/y). "atan" denotes the arctangent. Note that this
convention is not the one usually assumed in astronomy (i.e, we swap "x" by "y" above, or
similarly, ra => 90-ra, with respect to the astronomical convention)

```python
def find_3d_satellite_positions(sats, cen):
    # satellites:
    c1 = coord.ICRS(ra=sats['ra_gal']*u.degree, dec=sats['dec_gal']*u.degree, 
                    distance=(mice_cosmo.comoving_distance(sats['z_cgal'])))
    gc1 = c1.transform_to(coord.Galactocentric)
    # central
    c2 = coord.ICRS(ra=cen['ra_gal']*u.degree, dec=cen['dec_gal']*u.degree, 
                    distance=(mice_cosmo.comoving_distance(cen['z_cgal'])))
    gc2 = c2.transform_to(coord.Galactocentric)
    # sit on the central galaxy
    # xcen, ycen, zcen = 0., 0., 0.
    xsat, ysat, zsat = gc1.x-gc2.x, gc1.y-gc2.y, gc1.z-gc2.z 
    return xsat, ysat, zsat

def sat_rad_3d(xsat, ysat, zsat):
    return np.sqrt(xsat**2. + ysat**2. + zsat**2.)
```

```python
def satellite_shear_3d(r_rvir, A, b):
    return A*(r_rvir)**b

# find rvir
def rvir_halo(mice_cosmo, halo):
    rho_crit0 = mice_cosmo.critical_density0.to(u.M_sun*u.Mpc**(-3.))/(mice_cosmo.h**2.)
    rho_m = rho_crit0*mice_cosmo.Om0
    rho_halo = 200.*rho_m
    mass_halo = 10.**halo['lmhalo'][0]*u.M_sun # units of Msun/h 
    rvir = radvir(mass_halo, rho_halo)
    return rvir
```

```python
# measure a the average projected signal in bins of rp

def projected_signal_per_halo(rp_min, rp_max, nrpbins, rsat_proj, gamma_sat_proj):
    rp_histo = np.logspace(np.log10(rp_min.value),np.log10(rp_max.value), nrpbins+1) #*u.Mpc
    print rsat_proj
    gamma_proj_mean = np.zeros(nrpbins)
    err_gamma_proj = np.zeros(nrpbins)
    rp_bin = np.zeros(nrpbins)
    for ir in range(0,nrpbins):
        rp_bin[ir] = (rp_histo[ir]+0.5*(rp_histo[ir+1]-rp_histo[ir])) #.value
        bin_mask = np.where((rsat_proj >= rp_histo[ir]) & (rsat_proj < rp_histo[ir+1]))
        gamma_proj_mean[ir] = np.mean(gamma_sat_proj[bin_mask])
        err_gamma_proj[ir] = np.std(gamma_sat_proj[bin_mask])
    rp_bin = rp_bin #*u.Mpc

    return rp_bin, gamma_proj_mean, err_gamma_proj
```

```python
def projected_signal(rp_min, rp_max, nrpbins, rsat_proj, gamma_sat_proj):
    rp_histo = np.logspace(np.log10(rp_min),np.log10(rp_max), nrpbins+1) #*u.Mpc
    #print rp_histo
    #print rsat_proj
    gamma_proj_mean = np.zeros(nrpbins)
    err_gamma_proj = np.zeros(nrpbins)
    rp_bin = np.zeros(nrpbins)
    for ir in range(0,nrpbins):
        bin_mask = (rsat_proj >= rp_histo[ir]) & (rsat_proj < rp_histo[ir+1])
        #print rsat_proj[bin_mask]
        rp_bin[ir] = np.mean(rsat_proj[bin_mask].value) #(rp_histo[ir]+0.5*(rp_histo[ir+1]-rp_histo[ir])) #.value
        gamma_proj_mean[ir] = np.mean(gamma_sat_proj[bin_mask]).value
        err_gamma_proj[ir] = np.std(gamma_sat_proj[bin_mask]).value
    rp_bin = rp_bin #*u.Mpc

    return rp_bin, gamma_proj_mean, err_gamma_proj
```

```python
### Case 2: drawing from a cosine distribution

import numpy.random as random
from scipy.interpolate import interp1d 

def cdf_array(pdf, x):
    xmin = x.min()
    cdf = np.empty(len(x))
    for i in range (0,len(x)):
        cdf[i] = np.trapz(pdf[:i], x[:i])
    return cdf

def randoms_from_inverted_cdf(pdf, xvec, nrand):
    xmin = xvec.min()
    cdf = cdf_array(pdf, xvec)
    rdm = random.uniform(0,cdf.max(),nrand)
    num_inverse_cdf = interp1d(cdf, xvec)
    return num_inverse_cdf(rdm)

def cosine_pdf(x):
    norm = np.trapz(np.abs(np.cos(x)),x)
    #print('norm = ', norm)
    return np.cos(x)/norm

def generate_theta_randoms(theta_min, theta_max, nrandoms):
    theta_v = np.linspace(theta_min,theta_max, 500)
    cos_v = cosine_pdf(theta_v)
    random_theta_v = randoms_from_inverted_cdf(cos_v, theta_v, nrandoms)
    return random_theta_v
```

### What we have learnt

We performed a Monte Carlo simulation of a radial dependet satellite alignment distribution in 3D, assuming perfect alignment between the satellite major axes and the position of the central galaxy. We investigate how a power law radial dependence in 3D transforms when projected along the line of sight. We select 246 haloes containing more than 100 satellites each and assigned to each satellite a 3D shear $\bar{\gamma}(r)$ in the form of a power law. We assign random polar angles with respect to the line of sight and measured the projected shear in bins of projected separation $r_{sat} = r \sin \theta$.

We find that the slope of the power law is almost unchanged when going from 3D to 2D, while the amplitude is boosted by a factor $\pi/2$ with a 10$\%$ accuracy, compatible with the assumption that the projected signal is the integral over all of the possible configurations,

\begin{equation}
    \bar{\gamma}_{proj}(r_{proj}) \sim \int_0^{\pi} \bar{\gamma}(r) \sin^2 \theta d \theta
\end{equation}
