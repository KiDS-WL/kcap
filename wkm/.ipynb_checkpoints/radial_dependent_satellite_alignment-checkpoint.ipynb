```python
import matplotlib.colors as colors
import matplotlib.cm as cmx

def colormap_setup(NCURVES, my_colormap):
	values = range(NCURVES)
	my_cmap = cm = plt.get_cmap(my_colormap) 
	cNorm  = colors.Normalize(vmin=0, vmax=values[-1])
	scalarMap = cmx.ScalarMappable(norm=cNorm, cmap=my_cmap)
	colorVal = scalarMap.to_rgba(values)
	return colorVal
```

```python
def plot_setup(plot_type):
	# bigger figure for presentations:
	if plot_type == 'presentation':
		fig = plt.figure(figsize = (25,18))
		# fontsize = 'big'
		fs = 45
		ts = 20
		wt = 2.5
	elif plot_type == 'paper':
		# medium size for paper, with big labels:
		fig = plt.figure(figsize = (12,9))	
		# fontsize = 'medium'	
		fs = 40
		ts = 10
		wt = 1.5
	elif plot_type == 'notebook':
		fig = plt.figure(figsize = (11,7))	
		# fontsize = 'normal'			
		fs = 25
		ts = 10
		wt = 1.5		

	plt.rc('font',**{'family':'serif','serif':['Helvetica'], 'weight' : 'normal', 'size' : fs})	

	plt.rc('text', usetex=True)
	plt.rc('axes', linewidth=2)
		
	#if fontsize == 'big':
	plt.rc('xtick', labelsize=fs-5)
	plt.rc('xtick.major', size=ts, width=wt)
	plt.rc('xtick.minor', size=7, width=1.0)

	plt.rc('ytick', labelsize=fs-5)
	plt.rc('ytick.major', size=ts, width=wt)
	plt.rc('ytick.minor', size=7, width=1.0)
		
	return
```

# Compare with SB10

```python
def sb10_fell(theta_k, phi_k, l):
	phase = np.cos(2.*phi_k) + 1j*np.sin(2.*phi_k)
	sum1 = 0.
	for m in range(0,l+1):
		#print 'm = ', m
		#print 'newton_bin(l,m)*newton_bin(l+m-1,l) = ', newton_bin(l,m)*newton_bin(l+m-1,l)
		sum2=0.
		#print 'my coefficients = ', newton_bin(l,m)*newton_bin(l+m-1,l)
		for j in range(0,l-m+1):
			sum2 += binom(l-m,j) * g(j) * np.sin(theta_k)**(j) * np.cos(theta_k)**(l-m-j) * I_x(j+1,l-m-j)
			#print 'j = ', j
			#print 'sum2 = ', sum2
		sum1 += binom(l,m)*binom(0.5*(l+m-1.),l)*sum2		
	return 2.**l * sum1 #*phase 
```

```python
def wkm_SB10(uell, theta_k, phi_k, ell_max):
    nl = np.size(uell, axis=0)
    nz = np.size(uell, axis=1)
    nm = np.size(uell, axis=2)
    nk = np.size(uell, axis=3)
    #print 'nl, nz, nm, nk = ', nl, nz, nm, nk
    sum_ell = np.zeros([nz,nm,nk])
    for jz in range(nz):
        for im in range(nm):
            for ell in range(0,ell_max+1):
                radial = (1j)**(ell) * (2.*ell + 1.) * uell[ell,jz,im]
                angular = sb10_fell(theta_k, phi_k, ell)
                #print ell, angular
                sum_ell[jz,im] = sum_ell[jz,im] + radial*angular
                #print sum_ell
    return np.sqrt(np.real(sum_ell)**2.+np.imag(sum_ell)**2.)
```

```python
def setup(log_mass_min, log_mass_max, nmass, zmin, zmax, nz, kmin, kmax, nk):
    #log-spaced mass in units of M_sun/h
    mass = np.logspace(log_mass_min, log_mass_max, nmass)
    z_vec = np.linspace(zmin, zmax, nz)
    print z_vec
    k_vec = np.logspace(np.log10(kmin), np.log10(kmax), nk)
    return z_vec, nz, mass, nmass, k_vec, nk
```

```python
from astropy.cosmology import FlatLambdaCDM
def block_options(gamma_1h_amplitude, gamma_1h_slope):
    return gamma_1h_amplitude, gamma_1h_slope
    
def nfw_module(mass_in, k_in, z_in, ell_max_in):
    this_cosmo = FlatLambdaCDM(H0=70.0, Om0=0.25)
    rho_crit0 = this_cosmo.critical_density0.to(u.M_sun*u.Mpc**(-3.))/(this_cosmo.h**2.)
    mean_density0 = rho_crit0.value*this_cosmo.Om0
    #---- Comology ----#
    nmass = len(mass_in)
    nz = len(z_in)
    c_out = np.empty([nz,nmass])
    mass_term = 10.14/((mass_in/(2.e+12))**0.081)
    for jz in range (0,nz):
        for im in range (0,nmass):
            c_out[jz,im]= mass_term[im]/((1.+z_in[jz])**1.01)
    # compute the virial radius and the scale radius associated with a halo of mass M
    rho_m = mean_density0 #minimal_cosmo(omega_m)
    #print rho_m, minimal_cosmo(0.25)
    rho_halo = 200.*rho_m
    rvir_out = radvir(mass_in, rho_halo)
    r_s_out = np.empty([nz, nmass])
    for jz in range(0,nz):
        r_s_out[jz] = rvir_out/c_out[jz]
    return c_out, r_s_out, rvir_out
```

```python
def execute(config, block):
    z_setup, nz_setup, mass_setup, nmass_setup, k_setup, nk_setup = config[0], config[1], config[2], config[3], config[4], config[5] 
    #start_time = time.time()
    ell_max = 6
    # create intermediate variables to speed up the calculation
    nz = nz_setup
    z = z_setup
    #nmass = 100
    #mass = np.logspace(np.log10(mass_setup.min()), np.log10(mass_setup.max()), nmass)
    mass = mass_setup
    nk = 100
    k = np.logspace(-1, np.log10(k_setup.max()), nk)
    # Load slope of the power law that describes the satellite alignment
    g_1h_ampl, g_1h_slope = block[0], block[1]
    c_halo, r_s_halo, rvir_halo = nfw_module(mass, k, z, ell_max)   
    # uell[l,z,m,k]
    uell = IA_uell_gamma_r_hankel(g_1h_ampl, g_1h_slope, k, c_halo, z, r_s_halo, rvir_halo, mass, ell_max)	   
    # interpolate
    uell_interpolated = np.empty([ell_max+1, nz, nmass_setup, nk_setup])
    for il in range(0,ell_max+1):
        for jz in range(0,nz):
            f_interp = interp2d(k, mass, uell[il, jz], bounds_error=False, fill_value=0)
            uell_interpolated[il,jz] = f_interp(k_setup, mass_setup)            
    # wkm[nz,nmass,nk]
    theta_k = np.pi/2.
    phi_k = 0.
    wkm_out = wkm_my_fell(uell_interpolated, theta_k, phi_k, ell_max)   
    #print("--- wkm: %s seconds ---" % (time.time() - start_time))
    return z, mass_setup, k_setup, wkm_out
```

```python
from wkm_angular_part import another_fell

def wkm_my_fell(uell_in, theta_k, phi_k, ell_max):
    nl = np.size(uell_in, axis=0)
    nz = np.size(uell_in, axis=1)
    nm = np.size(uell_in, axis=2)
    nk = np.size(uell_in, axis=3)
    #print 'nl, nz, nm, nk = ', nl, nz, nm, nk
    sum_ell = np.zeros([nz,nm,nk])
    for jz in range(nz):
        for im in range(nm):
            for ell in range(0,ell_max+1):
                radial = (1j)**(ell) * (2.*ell + 1.) * uell_in[ell,jz,im]
                angular = another_fell(theta_k, phi_k, ell)
                #print ell, angular
                sum_ell[jz,im] = sum_ell[jz,im] + radial*angular
                #print sum_ell
    return np.sqrt(np.real(sum_ell)**2.+np.imag(sum_ell)**2.)
```

```python
def another_fell(theta_k, phi_k, l):
	phase = np.cos(2.*phi_k) + 1j*np.sin(2.*phi_k)
	sum1 = 0.
	for m in range(0,l+1):
		#print 'm = ', m
		#print 'newton_bin(l,m)*newton_bin(l+m-1,l) = ', newton_bin(l,m)*newton_bin(l+m-1,l)
		sum2=0.
		#print 'my coefficients = ', newton_bin(l,m)*newton_bin(l+m-1,l)
		for j in range(0,m+1):
			sum2 += binom(m,j) * g(m-j) * np.sin(theta_k)**(m-j) * np.cos(theta_k)**j * I_x(m-j+1,j)
			#print 'j = ', j
			#print 'sum2 = ', sum2
		sum1 += binom(l,m)*binom(0.5*(l+m-1.),l)*sum2		
	return 2.**l * sum1*phase 
```

```python
def wkm_test(uell, theta_k, phi_k, ell_max):
    nl = np.size(uell, axis=0)
    nz = np.size(uell, axis=1)
    nm = np.size(uell, axis=2)
    nk = np.size(uell, axis=3)
    #print 'nl, nz, nm, nk = ', nl, nz, nm, nk
    sum_ell = np.zeros([nz,nm,nk])
    for jz in range(nz):
        for im in range(nm):
            for ell in range(0,ell_max+1):
                radial = (1j)**(ell) * (2.*ell + 1.) * uell[ell,jz,im]
                angular = fell_analytic(theta_k, phi_k, ell)
                #print ell, angular
                sum_ell[jz,im] = sum_ell[jz,im] + radial*angular
                #print sum_ell
    return sum_ell
```
